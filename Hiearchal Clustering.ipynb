```r
library(factoextra)
```

```r
House_2012 <- read.csv("House_2012.csv")
House_Distances <- dist(House_2012[c(-1,-2,-3,-4)])
```

```r
HouseModel.single <-hclust(House_Distances,method="single")
plot(HouseModel.single)
```

```r
HouseModel.complete <-hclust(House_Distances,method="complete")
plot(HouseModel.complete)
```

```r
HouseModel.average <-hclust(House_Distances,method="average")
plot(HouseModel.average)
```

```r
HouseModel.single <-hclust(House_Distances,method="single")
HouseModel.single.clusters <- cutree(HouseModel.single, k = 4)
table(HouseModel.single.clusters,House_2012$party)
```

```r
House_2012[which(HouseModel.single.clusters==3),c(1,2,3,4)]
House_2012[which(HouseModel.single.clusters==4),c(1,2,3,4)]
```

```r
HouseModel.single.Cluster1 <- House_2012[which(HouseModel.single.clusters==1),c(1,2,3,4)]
HouseModel.single.Cluster1[which(HouseModel.single.Cluster1$party == 'R'),]
```

```r
HouseModel.complete <-hclust(House_Distances,method="complete")
HouseModel.complete.clusters <- cutree(HouseModel.complete, k = 2)
table(HouseModel.complete.clusters,House_2012$party)
```

```r
HouseModel.complete.Cluster1 <- House_2012[which(HouseModel.complete.clusters==1),c(1,2,3,4)]
HouseModel.complete.Cluster1[which(HouseModel.complete.Cluster1$party == 'R'),]
```

```r
HouseModel.average.clusters <- cutree(HouseModel.average, k = 2)
table(HouseModel.average.clusters,House_2012$party)
```

```r
HouseModel.average.Cluster1 <- House_2012[which(HouseModel.average.clusters==1),c(1,2,3,4)]
HouseModel.average.Cluster1[which(HouseModel.average.Cluster1$party == 'R'),]
```

```r
hc.complete.out=eclust(House_2012[c(-1,-2,-3,-4)], 
              FUNcluster = "hclust",
              hc_method="complete",
              k=2)
print(hc.complete.out$silinfo$avg.width)


hc.average.out=eclust(House_2012[c(-1,-2,-3,-4)], 
              FUNcluster = "hclust",
              hc_method="average",
              k=2,graph)
print(hc.average.out$silinfo$avg.width)

hc.single.out=eclust(House_2012[c(-1,-2,-3,-4)], 
              FUNcluster = "hclust",
              hc_method="single",
              k=4)
print(hc.single.out$silinfo$avg.width)
```
